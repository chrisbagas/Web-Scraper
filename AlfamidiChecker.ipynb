```python
import pandas as pd

df = pd.read_excel('ALFAMIDI_240917_Yogya.xlsx')
df.head()
```

```python
targets = ['clear sampo', 'sunsilk sampo', 'lifebuoy', 'tresemme shampoo', 'ponds',
           'glow &', 'vaseline', 'pepsodent pasta gigi',
           'closeup', 'lifebuoy sabun mandi', 'lux botanicals', 'rexona', 'axe', 'molto', 'sunlight',
           'wipol', 'vixal', 'royco', 'bango kecap', 'sariwangi', 'sarimurni', 'buavita',
           'head & shoulders sampo', 'pantene sampo', 'pantene shampoo', 'zinc', 'garnier', 'nivea',
           'marina', 'ciptadent pasta gigi', 'nuvo', 'giv', 'posh men body spray',
           'soklin softergent', 'soklin liquid', 'soklin pewangi', 'downy pelembut',
           'mama lemon', 'supersol', 'yuri porstex', 'masako penyedap rasa ayam',
           'sedaap kecap', 'abc kecap manis', 'sosro teh', 'sosro teh asli', 'tong tji', 'teh bendera',
           'country choice', 'citra', 'superpel']

# Group by location and identify missing products
missing_products_by_location = {}

for location, group in df.groupby('location'):
    available_products = set()
    
    # Check for each product in the group if it matches a target
    for product in group['productName']:
        for target in targets:
            if target.lower() in product.lower():
                available_products.add(target)
    
    # Find which targets are missing
    missing_products = set(targets) - available_products
    missing_products_by_location[location] = missing_products

# Print missing products by location
for location, missing in missing_products_by_location.items():
    print(f"Location: {location}")
    if missing:
        print("Missing products:")
        for product in missing:
            print(f" - {product}")
    else:
        print("All products available.")
    print()
```
