```python


check = ['royco', 'bango', 'sariwangi',
           'pepsodent', 'lifebuoy',
           'rexona', 'clear', 'sunsilk', 'glow',
           'sunlight', 'rinso', 'molto']

file_name = f"../Dagangan/DAGANGAN_{datetime.now().strftime('%y%m%d')}.xlsx"
# file_name = f"../Dagangan/DAGANGAN_250303.xlsx"
df = pd.read_excel(file_name)
df.head()
location_column = 'location'
# Loop through each unique location in the location column (if you have a column like that)
for location in df[location_column].unique():  
    print(f"\nChecking for missing products at location: {location}")
    
    # Filter the DataFrame for the current location 
    df_location = df[df[location_column] == location]
    
    # Convert 'productName' column to lowercase for case-insensitive matching
    product_names = df_location['productName'].str.lower()

    # Check for items in 'check' that are not found in any of the product names
    missing_in_df = [item for item in check if not product_names.str.contains(item).any()]

    # # Display the list of items from 'check' that are not in the DataFrame
    if missing_in_df:
        print("Items from the check list not found in the product names:")
        print(missing_in_df)
        dagangan_scrapper = DaganganData(
            '7.1.2', '127.0.0.1:5555', [location], missing_in_df)
        dagangan_scrapper.extractor(file_name)
    else:
        print("All items from the check list were found.")
```
